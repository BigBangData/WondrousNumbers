# Wondrous Numbers

```python
def next_step(x):
    if x % 2 == 0:
        return int(x/2)
    else:
        return x * 3 + 1
```

```python
def calc_seq(x):
    seq = []
    while x > 1:
        seq.append(next_step(x))
        x = next_step(x)
    return seq
```

```python
def get_seqs(seq):
    seqs = []
    for x in seq:
        seqs.append(calc_seq(x))
    return seqs
```

```python
seq = range(1, 10000)

seqs = get_seqs(seq)
```
